```python
# Plantilla de preprocesado

# Como importar las librerias

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

# Importamos el data set. Originalmente estaba en otra ruta pero por comodidad lo ponemos en el mismo path.
dataset = pd.read_csv('Salary_Data.csv')
```

```python
#TIPOS DE VARIABLES
# 1. CATEGORICAS: NOMINALES O ORDINALES
# LAS NOMINALES NO TIENEN NINGUN TIPO DE ORDEN: EJ HOMBRE/MUJER, COLORES, ETC (FACTORES)
# ORDINALES: TALLAS DE CAMISETA: S, M, L. NOTAS EEUU: A,B,C,D, ETC.
# 2. NUMERICAS: DISCRETAS
# LAS DISCRETAS: VARIABLES QUE PODEMOS CONTAR: EMPLEADOS DE UNA EMPRESA
# CONTINUAS: VARIABLES COMO ALTURA, PESO, QUE PUEDEN TOMAR VALORES DECIMALES.
```

```python
### REGRESION: PROCESO ESTADISTICO QUE SE BASA EN ESTIMAR RELACIONES EXISTENTES ENTRE VARIABLES.
### SE BASA EN ESTUDIAR EL COMPORTAMIENTO DE UNA VARIABLE DEPENDIENTE QUE DEPENDE DE VARIABLES "INDEPENDIENTES"
###  REGREGION LINEAL O LOGITSTICA
### REGRESION LINEAL SIMPLE: 1 VARIABLE INDEPENDIENTE
### REGRESION LINEAL MULTIPLE MULTIPLES VARIABLES INDEPENDIENTES.
### LO MISMO PARA LA LOGISTICA.

# REGRESION LINEAL: Y = B_0 + B_1*X
# LA VARIABLE INDEPENDIENTE ES X Y LA DEPENDIENTE ES Y. 
# B1 ES LA PENDIENTE DE LA RECTA = COEFICIENTE DEL MODELO DE REGRESION LINEAL
# B1 INDICA CUANTO DEBE INCREMENTARSE LA VARIABLE DEPENDIENTE POR CADA UNIDAD DE VARIABLE INDEPENDIENTE.
```

```python
# vamos a seleccionar las variables X (variables independientes)
# Matriz de variables independientes.
X = dataset.iloc[:,:-1].values # Seleccionamos los valores de todas las columnas menos la ultima.
# vamos a seleccionar la variables dependiente.
y = dataset.iloc[:,-1].values # seleccionamos los valores de la ultima columna
```

```python
# Dividir el data set en conjunto de entrenamiento y test
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=1/3,random_state=0) # el random state deber√≠a ser un numero aleatorio para hacerlo mas potente. Si queremos que siempre salga lo mismo,mantenemos el numero entero.
```

```python
### EN EL CASO DE LA REGRESION LINEAL SIMPLE NO HACE FALTA ESCALADO DE VARIABLES
### PORQUE AUTOMATICAMENTE EL MODELO LO HARA AUTOMATICAMENTE.
# from sklearn.preprocessing import StandardScaler
# sc_X = StandardScaler()
# X_train = sc_X.fit_transform(X_train)
# ## Se hace un escalado de estandarizacion. 
# ## Para hacerlo con la normalizacion se realiza un Normalizer.
# ## Se utiliza el Transform en el segundo caso para que lo haga en funcion del escalamiento de variables
# ## realizadas en el de entrenamiento.
# X_test = sc_X.transform(X_test)
```

```python
#### CREACION MODELO DE REGRESION LINEAL SIMPLE CON X_TRAIN Y Y_TRAIN
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
### SUMUNISTRAMOS AL MODEL LOS DATOS DE ENTRENAMIENTO: EL MODELO ESTA ENTRENADO.
regression.fit(X_train,y_train)
```

```python
### PREDECIR EL CONJUNTO DE TEST PARA VER SI EL MODELO SE AJUSTA.
y_pred = regression.predict(X_test)
```

```python
### VAMOS A PINTAR LOS DATOS PARA VER EN GRAFICO COMO SE AJUSTA EL MODELO
### VISUALIZACION DE RESULTADOS DE ENTRENAMIENTO
plt.scatter(X_train,y_train,color="red",label="y_train")
plt.scatter(X_train, regression.predict(X_train), color="black", label='y_pred')
plt.legend()
plt.plot(X_train,regression.predict(X_train),color = "blue")
plt.title("Sueldo vs a√±os de experiencia: Train Set")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Sueldo ($)")
plt.show()
```

```python
### REALMENTE EXISTE UNA CORRELACION LINEAL ENTRE LAS VARIABLES????
# VISUALIZACION DE LOS DATOS DE TEST
plt.scatter(X_test, y_test, color="red", label='y_pred')
# PODEMOS DEJAR X_TRAIN YA QUE CAMBIANDOLO POR EL TEST DARIA LA MISMA RECTA!!!
plt.plot(X_train,regression.predict(X_train),color = "blue")
plt.title("Sueldo vs a√±os de experiencia: Train Set")
plt.xlabel("A√±os de experiencia")
plt.ylabel("Sueldo ($)")
plt.show()
```
